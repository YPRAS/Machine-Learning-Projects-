### Handling Missing values 

Replacing all the null values of `Promo2`,`Promo2SinceWeek`,`Promo2SinceYear`,`PromoInterval` with 0.  

```python
merged_df = pd.read_csv('dataset/merged_data.csv')
missing_values = merged_df.isna().sum()
print(missing_values)
```

```python
missing_percentage = merged_df.isna().sum().div(merged_df.shape[0]).mul(100).to_frame().sort_values(by=0, ascending = False)
print(missing_percentage)
```

- Assuming  that missing values in the `CompetitionDistance` column mean there is no competition, filling them with zero. This indicates that the store has no nearby competitors.


- `CompetitionOpenSinceMonth` and `CompetitionOpenSinceYear`: Assuming Missing values could imply no competition was open before or that there was no competition.Replace missing values with 0 indicating no competition.
<br>

- CompetitionDistance is missing, but CompetitionOpenSinceMonth/Year is also missing:
<br>This suggests no competition exists for the store.
Filling both the CompetitionDistance and CompetitionOpenSinceMonth/Year with 0 to indicate there’s no competitor.

- CompetitionOpenSinceMonth/Year is missing, but CompetitionDistance is filled:
<br>This means a competitor exists, but the exact opening date wasn’t recorded.
In this case, filling CompetitionOpenSinceMonth/Year with 0 to indicate that the information was not recorded.

Conclusion:
Using 0 for missing values works as a best-case scenario for both conditions. It simplifies the handling of missing data while maintaining logical consistency across your dataset.

```python
# Fill missing CompetitionDistance with 0 (if competition does not exist)
# For rows where both CompetitionDistance and CompetitionOpenSinceMonth/Year are missing
merged_df[['CompetitionDistance', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear']] = merged_df[['CompetitionDistance', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear']].fillna(0)
print(merged_df.isna().sum())
```

### Handling Categorical Data

```python
# define numerical & categorical columns
numeric_features = [feature for feature in merged_df.columns if merged_df[feature].dtype != 'O']
categorical_features = [feature for feature in merged_df.columns if merged_df[feature].dtype == 'O']

print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```
